```python
# Function to get random images from a directory (reproducible & error-handling)
def get_random_images(directory, num_samples=5, seed=SEED):
    if not os.path.exists(directory):
        print(f"Warning: Directory '{directory}' not found.")
        return []

    image_files = [f for f in os.listdir(directory) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

    if len(image_files) == 0:
        print(f"Warning: No images found in {directory}.")
        return []

    random.seed(seed)  # Ensure reproducibility
    sample_files = random.sample(image_files, min(num_samples, len(image_files)))

    return [os.path.join(directory, file) for file in sample_files]

# Get random samples
fractured_samples = get_random_images(fractured_path, 5)
non_fractured_samples = get_random_images(non_fractured_path, 5)

# Display images
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

for i, img_path in enumerate(fractured_samples):
    try:
        img = Image.open(img_path)
        axes[0, i].imshow(img)
        axes[0, i].set_title("Fractured")
        axes[0, i].axis("off")
    except Exception as e:
        print(f"Error loading image: {img_path} - {e}")

for i, img_path in enumerate(non_fractured_samples):
    try:
        img = Image.open(img_path)
        axes[1, i].imshow(img)
        axes[1, i].set_title("Non-Fractured")
        axes[1, i].axis("off")
    except Exception as e:
        print(f"Error loading image: {img_path} - {e}")

plt.tight_layout()
plt.show()
```

# **Pre-Trained Models Baseline Accuracies**

Undersampling Non-Fractured images Due to Imbalance in Dataset

```python
# Function to get image paths
def get_image_paths(directory):
    return [os.path.join(directory, f) for f in os.listdir(directory) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

# Load image paths
fractured_images = get_image_paths(fractured_path)
non_fractured_images = get_image_paths(non_fractured_path)

# Undersampling Non-Fractured images to match Fractured images
non_fractured_sampled = random.sample(non_fractured_images, len(fractured_images))

# Create a balanced dataset
balanced_image_paths = fractured_images + non_fractured_sampled
balanced_labels = [0] * len(fractured_images) + [1] * len(non_fractured_sampled)

print(f"Balanced dataset size: {len(balanced_image_paths)} (Fractured: {len(fractured_images)}, Non-Fractured: {len(non_fractured_sampled)})")

# **Fix: Ensure stratified split is correct**
train_indices, val_indices = train_test_split(
    np.arange(len(balanced_image_paths)), test_size=0.2, random_state=SEED, stratify=balanced_labels
)

# Define dataset class
class BalancedDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        label = self.labels[idx]

        try:
            image = Image.open(image_path).convert('RGB')
        except Exception:
            #print(f"Skipping corrupted image: {image_path}")
            return self.__getitem__((idx + 1) % len(self.image_paths))  # Skip corrupt image

        if self.transform:
            image = self.transform(image)

        return image, label

# Define base transformation (Ensure it exists)
base_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Standard normalization
])

# Apply transformation to dataset
dataset = BalancedDataset(balanced_image_paths, balanced_labels, transform=base_transform)

# Create DataLoaders with Fixed Samplers
train_loader = DataLoader(dataset, batch_size=32, sampler=SubsetRandomSampler(train_indices))
val_loader = DataLoader(dataset, batch_size=32, sampler=SubsetRandomSampler(val_indices))

print(f"Train size: {len(train_indices)}, Validation size: {len(val_indices)}")
```

**Checking how many Grayscale vs. RGB images are in the data**

```python
import os
from PIL import Image

# Define paths to your datasets
fractured_path = '/content/drive/MyDrive/FracAtlas/FracAtlas/images/Fractured'
non_fractured_path = '/content/drive/MyDrive/FracAtlas/FracAtlas/images/Non_fractured'

# Function to count grayscale vs. RGB images
def count_image_modes(directory):
    grayscale_count = 0
    rgb_count = 0
    other_count = 0  # If images have unexpected formats

    for file in os.listdir(directory):
        if file.lower().endswith(('.jpg', '.jpeg', '.png')):
            try:
                img = Image.open(os.path.join(directory, file))
                if img.mode == "L":  # Grayscale
                    grayscale_count += 1
                elif img.mode == "RGB":  # RGB
                    rgb_count += 1
                else:
                    other_count += 1  # Other modes (CMYK, RGBA, etc.)
            except Exception as e:
                print(f"Skipping corrupted image: {file} - {e}")

    return grayscale_count, rgb_count, other_count

# Count images in each folder
fractured_gray, fractured_rgb, fractured_other = count_image_modes(fractured_path)
non_fractured_gray, non_fractured_rgb, non_fractured_other = count_image_modes(non_fractured_path)

# Print results
print("Fractured Images:")
print(f"  Grayscale: {fractured_gray}")
print(f"  RGB: {fractured_rgb}")
print(f"  Other: {fractured_other}\n")

print("Non-Fractured Images:")
print(f"  Grayscale: {non_fractured_gray}")
print(f"  RGB: {non_fractured_rgb}")
print(f"  Other: {non_fractured_other}\n")

# Total counts
total_gray = fractured_gray + non_fractured_gray
total_rgb = fractured_rgb + non_fractured_rgb
total_other = fractured_other + non_fractured_other

print("Total Dataset Summary:")
print(f"  Grayscale: {total_gray}")
print(f"  RGB: {total_rgb}")
print(f"  Other: {total_other}")

# Determine which is dominant
if total_rgb > total_gray:
    print("\n🔹 Your dataset is mostly **RGB** images.")
elif total_gray > total_rgb:
    print("\n🔹 Your dataset is mostly **Grayscale** images.")
else:
    print("\n🔹 Your dataset has an equal number of **RGB** and **Grayscale** images.")
```

**Converting Grayscale Images to RGB**

```python
class FixedDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        label = self.labels[idx]

        try:
            image = Image.open(image_path)
            if image.mode != "RGB":
                image = image.convert("RGB")  # Convert grayscale to RGB
        except Exception:
            #print(f"Skipping corrupted image: {image_path}")
            return self.__getitem__((idx + 1) % len(self.image_paths))  # Skip corrupt image

        if self.transform:
            image = self.transform(image)

        return image, label
```

Testing Different Pre-Trained Models (lr = 0.001)

```python
# Set device (GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define pretrained models to test
model_list = {
    "ResNet18": models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1),
    "ResNet50": models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1),
    "EfficientNet_B0": models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1),
    "MobileNetV2": models.mobilenet_v2(weights=models.MobileNet_V2_Weights.IMAGENET1K_V1),
}

# Function to modify the classifier for binary classification
def modify_classifier(model):
    if isinstance(model, models.ResNet) or isinstance(model, models.DenseNet):
        model.fc = nn.Linear(model.fc.in_features, 2)  # Change output layer to 2 classes
    elif isinstance(model, models.EfficientNet):
        model.classifier[1] = nn.Linear(model.classifier[1].in_features, 2)
    elif isinstance(model, models.MobileNetV2):
        model.classifier[1] = nn.Linear(model.classifier[1].in_features, 2)
    return model

# Ensure reproducibility in train/val split
SEED = 42
torch.manual_seed(SEED)

# Train/Validation Split (Fixed)
train_indices, val_indices = train_test_split(
    list(range(len(balanced_image_paths))), test_size=0.2, random_state=SEED, stratify=balanced_labels
)

# Create DataLoaders with Fixed Subsets
train_subset = Subset(dataset, train_indices)
val_subset = Subset(dataset, val_indices)

train_loader = DataLoader(train_subset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_subset, batch_size=32, shuffle=False)

# Dictionary to store results
results = {}

for model_name, model in model_list.items():
    print(f"\nTraining {model_name}...")

    # Modify model for 2-class classification
    model = modify_classifier(model).to(device)

    # Define loss function & optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Train for a few epochs
    num_epochs = 10
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct, total = 0, 0

        for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}"):
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

        train_acc = 100 * correct / total
        print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss / len(train_loader):.4f}, Train Accuracy: {train_acc:.2f}%")

    # Validation
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

    accuracy = 100 * correct / total
    results[model_name] = accuracy
    print(f"Validation Accuracy for {model_name}: {accuracy:.2f}%")

# Print all results
print("\nFinal Pretrained Model Performance:")
for model, acc in results.items():
    print(f"{model}: {acc:.2f}%")
```

Modified Dataset Class for Transformation Testing

```python
# Custom Dataset with Weighted Classes
class WeightedTransformedDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None, extra_transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform
        self.extra_transform = extra_transform  # Choose transformation (Histogram Eq, CLAHE, UM)

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        label = self.labels[idx]

        try:
            image = Image.open(image_path).convert('RGB')
        except Exception:
            print(f"Skipping corrupted image: {image_path}")
            return self.__getitem__((idx + 1) % len(self.image_paths))  # Skip corrupt image

        if self.extra_transform:
            image = self.extra_transform(image)  # Apply selected transformation

        if self.transform:
            image = self.transform(image)

        return image, label

# Compute class weights for imbalanced data
def compute_class_weights(labels, device):
    num_fractured = sum(1 for label in labels if label == 0)
    num_non_fractured = sum(1 for label in labels if label == 1)
    total_samples = num_fractured + num_non_fractured

    fractured_weight = total_samples / (2 * num_fractured)
    non_fractured_weight = total_samples / (2 * num_non_fractured)

    return torch.tensor([fractured_weight, non_fractured_weight]).to(device)
```

Unsharp Masking (UM)

```python
# Define Unsharp Masking Transformation
def unsharp_masking(image, radius=2, amount=1.5):
    """Applies Unsharp Masking to the image."""
    img_cv = np.array(image)
    blurred = cv2.GaussianBlur(img_cv, (0, 0), radius)  # Apply Gaussian blur
    img_um = cv2.addWeighted(img_cv, 1 + amount, blurred, -amount, 0)  # Unsharp mask
    return Image.fromarray(img_um)  # Convert back to PIL format

# Apply Unsharp Masking
train_dataset = WeightedTransformedDataset(balanced_image_paths, balanced_labels, transform=base_transform, extra_transform=unsharp_masking)

# Compute class weights
class_weights = compute_class_weights(full_labels, device)

# Use Weighted Sampler
num_samples = len(full_labels)
sample_weights = [class_weights[label].item() for label in full_labels]
sampler = WeightedRandomSampler(sample_weights, num_samples, replacement=True)

# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=32, sampler=sampler)

# Validation dataset (NO transformations)
val_dataset = WeightedTransformedDataset(full_image_paths, full_labels, transform=base_transform, extra_transform=None)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

# Train Model with Unsharp Masking
print("\nTraining EfficientNet_B0 with Unsharp Masking...")
```
